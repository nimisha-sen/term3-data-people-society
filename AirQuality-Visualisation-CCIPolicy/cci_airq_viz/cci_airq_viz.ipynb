```python
import pandas as pd
import json
import matplotlib.pyplot as plt

# Load the data from the CSV file
file_path = '/Users/nimishasen/Desktop/cci_airq_viz/log_data.csv' 
data = pd.read_csv(file_path)

# Parse the JSON log messages into a list of dictionaries
logs = data['log_message'].apply(json.loads)

# Convert the list of dictionaries into a DataFrame for easier analysis
df = pd.json_normalize(logs)

# Convert the 'time' column to datetime format
df['time'] = pd.to_datetime(df['time'])

# Plot particulate matter (PM0.3, PM1.0, PM2.0, and PM10) over time
plt.figure(figsize=(14, 7))
plt.plot(df['time'], df['message.pm003_count'], color='blue', label='PM0.3 Count')
plt.plot(df['time'], df['message.pm01'], color='green', label='PM1.0 (µg/m³)')
plt.plot(df['time'], df['message.pm02'], color='#E57373', label='PM2.0 (µg/m³)')
plt.plot(df['time'], df['message.pm10'], color='purple', label='PM10 (µg/m³)')
plt.xlabel('Time')
plt.ylabel('Concentration / Count')
plt.title('Particulate Matter (PM0.3, PM1.0, PM2.0, and PM10) Over Time')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(14, 7))
plt.plot(df['time'], df['message.rco2'], label='CO2 (ppm)')
plt.xlabel('Time')
plt.ylabel('CO2 (ppm)')
plt.title('CO2 Levels Over Time')
plt.legend()
plt.grid(True)
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(14, 7))

ax1.set_xlabel('Time')
ax1.set_ylabel('Temperature (°C)', color='tab:blue')
ax1.plot(df['time'], df['message.atmp'], label='Temperature (°C)', color='tab:blue')
ax1.tick_params(axis='y', labelcolor='tab:blue')

ax2 = ax1.twinx()
ax2.set_ylabel('Relative Humidity (%)', color='tab:green')
ax2.plot(df['time'], df['message.rhum'], label='Relative Humidity (%)', color='tab:green')
ax2.tick_params(axis='y', labelcolor='tab:green')

fig.tight_layout()
plt.title('Temperature and Relative Humidity Over Time')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(14, 7))
plt.plot(df['time'], df['message.tvoc_index'], label='TVOC Index')
plt.plot(df['time'], df['message.nox_index'], label='NOx Index')
plt.xlabel('Time')
plt.ylabel('Index Value')
plt.title('TVOC and NOx Index Over Time')
plt.legend()
plt.grid(True)
plt.show()
```
